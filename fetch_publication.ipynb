```python
#fetch author name
def print_author_name(aid):
    url = f"https://api.openalex.org/authors/{aid}"
    response = requests.get(url)
    author_data = response.json()

    print(f"Name: {author_data.get('display_name')}")

print_author_name(aid)
```

```python
def build_author_works_url_aid(aid, page=1, per_page=200):
    endpoint = 'works'
    # filters = (
    #     f'author.orcid:{id}',
    #     'is_paratext:false'
    # )
    # Include pagination parameters
    url = f"https://api.openalex.org/works?filter=author.id:{aid}&per_page={per_page}&page={page}"
    return url
```

```python
def fetch_all_author_works(aid):
    all_works = []
    page = 1
    per_page = 200  # Max allowed by OpenAlex

    while True:
        # Build URL for the current page
        url = build_author_works_url_aid(aid, page=page, per_page=per_page)
        response = requests.get(url)
        
        if response.status_code != 200:
            print(f"Error fetching page {page}: {response.status_code}")
            break
        
        data = response.json()
        works = data.get('results', [])
        all_works.extend(works)
        
        # Check if we've fetched all results
        meta = data.get('meta', {})
        total_pages = meta.get('count', 0) // per_page + 1
        
        print(f"Fetched page {page}/{total_pages} ({len(works)} works)")
        
        if page >= total_pages:
            break
        page += 1
        
    return all_works
```

```python
def build_author_works_url_orchid(orcid, page=1, per_page=200):
    endpoint = 'works'
    filters = (
        f'author.orcid:{orcid}',
        'is_paratext:false'
    )
    # Include pagination parameters
    url = f'https://api.openalex.org/{endpoint}?filter={",".join(filters)}&per_page={per_page}&page={page}'
    return url
```
